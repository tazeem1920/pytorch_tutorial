#Intitilaizign a neural network

```python
import torch
import torch.nn as nn
import torch.nn.functional as F
```

```python
# create a model class that inherits nn.Module
class Model(nn.Module):
  #input layers (number of features)--> hidden 1--> hidden2 ---> output layer
  def __init__(self, in_features=4,h1=8,h2=9,out_features=3):
    super().__init__()
    self.fc1=nn.Linear(in_features,h1)
    self.fc2=nn.Linear(h1,h2)
    self.out=nn.Linear(h2,out_features)

  def forward(self, x):
    x=F.relu(self.fc1(x))
    x=F.relu(self.fc2(x))
    x=self.out(x)

    return x

```

```python
torch.manual_seed(41)
```

```python
#  creataing an instance of model
model=Model()
```

# data input and training the model

```python
import pandas as pd
```

```python
url='https://gist.githubusercontent.com/curran/a08a1080b88344b0c8a7/raw/0e7a9b0a5d22642a06d3d5b9bcbad9890c8ee534/iris.csv'
my_df=pd.read_csv(url)
```

```python
my_df.tail()
```

```python
my_df.iloc[:,-1]
```

```python
my_df.iloc[:,-1].unique()
```

```python
my_df['species']=my_df['species'].replace('setosa',0.0)
my_df['species']=my_df['species'].replace('versicolor',1.0)
my_df['species']=my_df['species'].replace('virginica',2.0)
```

```python
my_df
```

# train test spllit

```python
X=my_df.drop('species',axis=1)
y=my_df['species']
```

```python
# convert these to numpy array for better calcultions
X=X.values
y=y.values
```

```python
# splitting
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test= train_test_split(X,y, test_size=0.2, random_state=20)
```

```python
X_train=torch.FloatTensor(X_train)
X_test=torch.FloatTensor(X_test)
y_train=torch.LongTensor(y_train)
y_test=torch.LongTensor(y_test)
```

#setting the criteria for model training.

```python
criteraion=nn.CrossEntropyLoss()
optimizer=torch.optim.Adam(model.parameters(),lr=0.01)
```

```python
model.parameters()
```

```python
params = list(model.parameters())
print(params)
```

```python
for param in model.parameters():
    print(param.shape)
```

```python
epochs=100
losses=[]
for i in range(epochs):

  #go forwared and get a predictio
  y_pred=model.forward(X_train)


  # measure the losss
  loss=criteraion(y_pred,y_train)
  losses.append(loss.detach().numpy())

  #print every 10 epochs
  if i%10==0:
    print(f'Epoch: {i} and loss: {loss}')
  optimizer.zero_grad()
  loss.backward()
  optimizer.step()
```

```python
print(losses)
```

```python
from matplotlib import pyplot as plt
```

```python
# graph the loss
plt.plot(range(epochs),losses)
plt.ylabel("loss/error")
plt.xlabel("epochs")
```

```python
# Evaluate Model on Test Data Set (Validate model on test set)
with torch.no_grad():  # Basically turn off the back propagation
  y_eval=model.forward(X_test)
  loss=criteraion(y_eval,y_test)

```

```python
loss
```

```python
correct=0
with torch.no_grad():
  for i, data in enumerate (X_test):
    y_val=model.forward(data)

    if y_test[i]==0:
      x="sentosa"
    elif y_test[i]==1:
      x="versicolor"
    else:
      x="virginica"

    print(f'{i+1}.)  {str(y_val)} \t{y_test[i]} \t{y_val.argmax().item()} \t {x}')

# correct or not
    if y_val.argmax().item() == y_test[i]:
      correct+=1
print(f'we got {correct} correct out of {len(y_test)}')
```

```python
accuracy=correct/y_test.size(0)
print(f'Accuracy : {accuracy}')
```
